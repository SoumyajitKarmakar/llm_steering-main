```python
# check unique

def check_dup(file_name):

    if isinstance(file_name, str):
        data = []
        with open(file_name, "r") as f:
            lines = f.readlines()
            for line in lines:
                data.append(line.strip())
    elif isinstance(file_name, list):
        data = file_name
    else:
        return

    duplicates = {}
    for i in data:
        c = data.count(i)
        if c > 1:
            # duplicates.append(i)
            duplicates[i] = c

    return duplicates
```

```python
def read_to_list(file_name):
    data = []
    with open(file_name, "r") as f:
        lines = f.readlines()
        for line in lines:
            data.append(line.strip())
    
    return data
```
