```python
# check unique

def check_dup(file_name):

    if isinstance(file_name, str):
        data = []
        with open(file_name, "r") as f:
            lines = f.readlines()
            for line in lines:
                data.append(line.strip())
    elif isinstance(file_name, list):
        data = file_name
    else:
        return

    duplicates = {}
    for i in data:
        c = data.count(i)
        if c > 1:
            # duplicates.append(i)
            duplicates[i] = c

    return duplicates
```

```python
def read_to_list(file_name):
    data = []
    with open(file_name, "r") as f:
        lines = f.readlines()
        for line in lines:
            data.append(line.strip())
    
    return data
```

```python
file_name = "g_moods/moods_antonyms.txt"
```

```python
l = read_to_list(file_name)
```

```python
dups = dict(sorted(check_dup(l).items(), key=lambda item: item[1], reverse=True))
print(dups)
print(len(dups))
```
