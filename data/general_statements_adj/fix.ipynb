```python
def clean(input_filename, output_filename):
    with open(input_filename, 'r') as infile, open(output_filename, 'w') as outfile:
        for line in infile:
            cleaned_line = line.rstrip()
            if cleaned_line == "":
                continue
        
            outfile.write(cleaned_line + '\n') # Add a single newline back
```

```python
input_filename0 = '{}/class_0_old.txt'
output_filename0 = '{}/class_0.txt'

input_filename1 = '{}/class_1_old.txt'
output_filename1 = '{}/class_1.txt'

for fol in ["complexity", "logic", "physical", "social", "state", "texture", "time"]:
    clean(input_filename0.format(fol), output_filename0.format(fol))
    clean(input_filename1.format(fol), output_filename1.format(fol))
```
