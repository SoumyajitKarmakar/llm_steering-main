```python
# check unique

def check_dup(data):
    duplicates = []
    for i in data:
        if data.count(i) > 1:
            duplicates.append(i)
    return duplicates
```

```python
total_list = []
with open(file_name, "r") as f:
    lines = f.readlines()
    for line in lines:
        total_list.append(line.strip())
```

```python
print(len(total_list)) # 956
dup = check_dup(total_list)
print(dup)
print(len(dup))
```
