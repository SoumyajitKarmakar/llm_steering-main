```python
# check unique

def check_dup(data):
    duplicates = []
    for i in data:
        if data.count(i) > 1:
            duplicates.append(i)
    return duplicates
```

```python
print(len(first))
dup = check_dup(first)
print(dup)
print(len(dup))
```

```python
print(len(all_file))

all_clean = set([str(i).split("/")[-1].split("_")[1] for i in all_file])
```

```python
print(all_clean)
print(len(all_clean))
```

```python
left = uniq - all_clean
```

```python
print(len(uniq))
print(len(all_clean))
print(len(left))
```
