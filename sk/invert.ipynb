```python
X_test, Y_test = read_tuples_as_list(llm, test_data, path='../directions_moods_plus_llama/')
```

```python
def calc_val_loss(X_test, Y_test, lrr_matrix):
    avg_loss = []

    for i in range(len(X_test)):
        y_pred = apply_trans(X_test[i], lrr_matrix)

        loss_list = []
        for layer in y_pred:
            layer_loss = torch.mean(torch.norm(y_pred[layer] - Y_test[i][layer], dim=1))

            loss_list.append(layer_loss)

        avg_loss.append(sum(loss_list) / len(loss_list))
        
    avg_val_loss = sum(avg_loss) / len(avg_loss)

    return avg_val_loss
```

```python
def calc_involution_error(lrr_matrix):
    involution_error_list = []

    for i in lrr_matrix:
        layer_involution_error = torch.norm(lrr_matrix[i] @ lrr_matrix[i] - torch.eye(lrr_matrix[i].shape[0]).to("cuda"), p='fro')
        involution_error_list.append(layer_involution_error)

    avg_involution_error = sum(involution_error_list) / len(involution_error_list)

    return avg_involution_error
```

```python
print("Val loss 1 way" , calc_val_loss(X_test, Y_test, lrr_matrix_1way))
print("Involution error 1 way" , calc_involution_error(lrr_matrix_1way))
```

```python
print("Val loss 2 way" , calc_val_loss(X_test, Y_test, lrr_matrix_2way))
print("Involution error 2 way" , calc_involution_error(lrr_matrix_2way))
```
